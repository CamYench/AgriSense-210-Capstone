```python
import boto3

# Initialize S3 client
s3_client = boto3.client('s3')

# S3 bucket name
bucket_name = 'agrisense'

def list_s3_bucket_contents(bucket):
    """List and print the contents of an S3 bucket

    :param bucket: Bucket to list contents of
    """
    try:
        paginator = s3_client.get_paginator('list_objects_v2')
        pages = paginator.paginate(Bucket=bucket)

        print(f"Contents of bucket '{bucket}':")
        for page in pages:
            if 'Contents' in page:
                for obj in page['Contents']:
                    print(obj['Key'])
            else:
                print(f"No contents found in page {page}")
    except Exception as e:
        print(f"Error listing contents of bucket '{bucket}': {str(e)}")

# List and print contents of the S3 bucket
list_s3_bucket_contents(bucket_name)
```
